```python
from urllib.request import urlopen
import re
```

```python
url= "https://www.farandwide.com/s/most-visited-cities-world-0b1e6081140746e3"
page = urlopen(url)
```

```python
page = page.read()
page = page.decode("utf-8")
print(page)
```

```python
fp = open("Cities2019.html", "w")
fp.write(page)
fp.close()

start = page.find("<article>")
end = page.find("</article>")
print(start, end)

f = open("Cities2019.html", "w")
relevant_text = page[start:end]
f.write(relevant_text)
f.close()
```

```python
relevant_text = re.sub("\s+", "", relevant_text) 
#print(relevant_text)
newList1 = re.findall("<h2.*?>.*?</h2>", relevant_text, re.DOTALL)
newList2 = re.findall("Arrivals:.*?</p>", relevant_text, re.DOTALL)
print(newList2)
iter = 0
for i in newList1:
    elem = re.sub("<h2>","", i)
    elem = re.sub("</h2>", ",", elem)
    elem = re.sub(":", ",", elem)
    newList1[iter] = elem
    iter+=1
    #print(elem)
newList1 = newList1[1:-1]

fp = open("Cities2019.csv", "w")
for i in range(len(newList1)):
    elem = re.sub("Arrivals:.*?</strong>", "", newList2[i])
    elem = re.sub("</p>", "", elem)
    fp.write(newList1[i])
    fp.write(elem)
    fp.write("\n")
fp.close()
print(newList2)
```

```python
fp = open("Cities2019.csv", "r")
string = fp.read()
#print(string)
fp.close()
```

```python
cityList = re.split("\n", string)
cityList = cityList[:-1]
print(cityList)

iter = 0
for i in cityList:
    tempList = re.split(",", i)
    tourists = tempList[-1]
    tourists = re.split("m|&", tourists)
    tourists = float(tourists[0])
    tourists *= 1000000
    #print(tourists)
    tempList[-1] = str(tourists)
    cityList[iter] = ','.join(tempList)
    iter+=1

print(cityList)
```

```python
iter = len(cityList)-1
f = open("Cities2019-1.csv", "w+")
while (iter>=0):
    f.write(cityList[iter])
    f.write("\n")
    iter-=1
```
