```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "http://www.formula1-dictionary.net/driver_constructor_champ.html"
```

```python
page = urlopen(url)
page = page.read()
page = page.decode("utf-8")
print(page)
```

```python
with open("newFile.html", "w", encoding='utf-8') as f:
    f.write(page)
f.close()
```

```python
import re

table = re.findall("<table.*?>.*?</table>", page, re.DOTALL)
champs = table[0]
records = table[1]
print(champs)
```
